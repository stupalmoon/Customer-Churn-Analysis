```python
# Resetting from original data or reloading the file
df = pd.read_csv("C:/Users/stupa/Downloads/Customer Churn.csv")  # Reload clean version
# Now do conversion
df["SeniorCitizen"] = df["SeniorCitizen"].astype(int)
df["SeniorCitizen"] = df["SeniorCitizen"].apply(lambda x: "yes" if x == 1 else "no")

```

```python
def conv(value):
    try:
        return "yes" if int(value) == 1 else "no"
    except:
        return value  # Leave it unchanged if not convertible

df["SeniorCitizen"] = df["SeniorCitizen"].apply(conv)

```

```python
columns = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

# Set up subplot grid (3 rows, 3 columns for 9 plots)
fig, axes = plt.subplots(3, 3, figsize=(15, 12))
axes = axes.flatten()  # flatten to make indexing easier

# Create one plot per column
for i, col in enumerate(columns):
    sns.countplot(x=col, data=df, hue=df["Churn"], ax=axes[i])
    axes[i].set_title(f'Count of {col}')
    axes[i].tick_params(axis='x', rotation=0)

plt.tight_layout()
plt.show()
```
